```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from math import sqrt


# drop the first column 
# x = the data frame with all the rows and all the columns except the first one
# y = the target column
x = data.iloc[:, 1:-1]
y = data['target']

#x_train = the data we use to train the model
#x_test = the data we use to test the model
#y_train = the target values we use to train the model
#y_test = the target values we use to test the model

# test_size = 0.2 means that 20% of the data will be used for testing and 80% will be used for training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=69)

numNeighbors = int(sqrt(len(x_train))) - 31
print(f'numNeighbors: {numNeighbors}')

# Create a KNN classifier with the number of neighbors
knn = KNeighborsClassifier(n_neighbors=numNeighbors, weights='distance', algorithm='auto', metric='euclidean')

# ball_tree : accuracy: 24.21%, took 17.6s to run
# kd_tree : accuracy: 24.21%, took 22.2s to run
# brute : accuracy: 23.97%, took 2.0s to run
# auto : accuracy: 23.97%, took 1.4s to run

# numNeighbors : 40, accuracy: 24.21%, took 1.4s to run
# numNeighbors : 39, accuracy 23.97% took 1.4s to run
# numNeighbors : 29, accuracy 23.97% took 1.3s to run
# numNeighbors : 19, accuracy 24.46%, took 1.3s to run
# numNeighbors : 9, accuracy 27.36%, took 1.4s to run 
# numNeighbors : 7, accuracy 18.40!, took 1.2s to run 
# numNeighbors : 5, accuracy 22.28%, took 1.2s to run
# numNeighbors : 3, accuracy 24.29%, took 1.4s to run
# numNeighbors : 1, accuracy 21.31%, took 1.3s to run


# Train the classifier
knn.fit(x_train, y_train)

# Predict the labels for the test set
y_pred = knn.predict(x_test)

# Print the accuracy of the classifier as a percentage
accuracy = knn.score(x_test, y_test) * 100
print(f'Accuracy: {accuracy:.2f}%')
```
