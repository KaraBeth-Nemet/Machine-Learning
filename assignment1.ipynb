```python
# importing libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

iris = load_iris()  # loading the iris dataset

data = pd.DataFrame(iris.data, columns=iris.feature_names)

data['class'] = iris.target_names[iris.target]

print(data)
```

```python
# find the number of each class in the dataset
print(data['class'].value_counts())


nice_data = pd.DataFrame(data, columns=['class', 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])

display(nice_data)

# find average of sepal length

print ("Average of sepal length : ", nice_data['sepal length (cm)'].mean(), "cm")

# find average of sepal width

print ("Average of sepal width : ", nice_data['sepal width (cm)'].mean(), "cm") 

# find average of petal length

print ("Average of petal length : ", nice_data['petal length (cm)'].mean(), "cm")

# find average of petal width

print ("Average of petal width : ", nice_data['petal width (cm)'].mean(), "cm")



```

```python
# create a pie chart with seaborn / matplotlib

data_class = data['class'].value_counts()
keys = data['class'].value_counts().keys()

# declaring exploding pie 
explode = [0, 0.1, 0] 
# define Seaborn color palette to use 
palette_color = sns.color_palette('deep') 
  
# plotting data on chart 
plt.pie(data_class, labels=keys, colors=palette_color, 
        explode=explode, autopct='%.0f%%') 
  
# displaying chart 
plt.show() 
```
