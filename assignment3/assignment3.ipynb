```python
import pandas as pd

data = pd.read_csv('dataset.csv')

display(data)

# print out all the names of the columns
print(data.columns)
```

```python
# og number of rows 
print(data.shape)

# drop NAN values
# Pandas function : "Drop the rows where at least one element is missing."
data = data.dropna()
#print data sizes so we can see if we dropped any rows 
print(data.shape)

# drop duplicates
# Pandas function : "By default, it removes duplicate rows based on all columns."
data = data.drop_duplicates()
# print data sizes so we can see if we dropped any rows
print(data.shape)
```

```python

# make a target column in the original data frame with a target value of 0 for breast cancer, 1 for prostate cancer, and 2 for lung cancer
# why do we need a target column?
# We need a target column to be able to train the model to predict the target value. 
# The target value is the value we are trying to predict.

# set every tartget value to 0
data['target'] = 0
# change the target value to 1 if the first column starts with 'prad'
data.loc[data.iloc[:, 0].str.startswith('prad'), 'target'] = 1
# change the target value to 2 if the first column starts with 'luad'
data.loc[data.iloc[:, 0].str.startswith('luad'), 'target'] = 2

display(data)
```

```python

# drop the first column 
# x = the data frame with all the rows and all the columns except the first one
# y = the target column
x = data.iloc[:, 1:-1]
y = data['target']

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

#x_train = the data we use to train the model
#x_test = the data we use to test the model
#y_train = the target values we use to train the model
#y_test = the target values we use to test the model

# test_size = 0.2 means that 20% of the data will be used for testing and 80% will be used for training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# create a logistic regression model
model = LogisticRegression(max_iter=100)

# train the model
model.fit(x_train, y_train)

# predict the target values for the test data
y_pred = model.predict(x_test)

from sklearn.metrics import accuracy_score
# accuracy_score = the number of correct predictions / the number of all predictions
print("Accuracy : ")
print(accuracy_score(y_test, y_pred))

# print precision 
from sklearn.metrics import precision_score
# percision = the number of correct positive predictions / the number of all positive predictions
# percision = TP / (TP + FP)
print("Precision : ")
print(precision_score(y_test, y_pred, average=None))

# print recall
from sklearn.metrics import recall_score
# recall = the number of correct positive predictions / the number of positive predictions that should have been made
# recall = TP / (TP + FN)
print("Recall : ")
print(recall_score(y_test, y_pred, average=None))

#print classification report
from sklearn.metrics import classification_report
print("Classification Report : ")
print(classification_report(y_test, y_pred))


```
