- What are perceptrons?
    - Perceptron is a type of neural network that performs binary classification that maps input features to an output decision, usually classifying data into one of two categories, such as 0 or 1.
<br>
<br>
- What are the main steps of perceptrons?
    - Step 1: Initialize the weight and learning rate
        - We consider the weight values for the number of inputs + 1 (with the additional +1 accounting for the bias term
    - Step 2: Define the Linear Layer
        - The first step is to calculate the weighted sum of the inputs. This is done using the formula: Z = XW + b, where X represents the inputs, W the weights, and b the bias.
    - Step 3: Define the Activation Function
        - The Heaviside Step function is used as the activation function, which compares the weighted sum to a threshold. If the sum is greater than or equal to 0, it outputs 1; otherwise, it outputs 0.
    - Step 4: Define the Prediction
        - Use the linear function followed by the activation function to generate predictions based on the input features.
    - Step 5: Define the Loss Function
        - The loss function calculates the error between the predicted output and the actual output. In the Perceptron, the loss is the difference between the target value and the predicted value.
    - Step 6: Define Training
    - Step 7: Fit the Model

- How can perceptrons be used to identify groups of cancer patients in the dataset?
    - Perceptrons can be used to classify patients into different groups based on their features (e.g., age, tumor size, genetic markers). By training the perceptron on labeled data (where the groups are known), it can learn to predict the group of new, unseen patients. This can help in identifying high-risk patients and tailoring personalized treatment plans.


```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('dataset.csv')

data = data.dropna()
data = data.drop_duplicates()

# set every tartget value to 0
data['target'] = 0
# change the target value to 1 if the first column starts with 'prad'
data.loc[data.iloc[:, 0].str.startswith('prad'), 'target'] = 1
# change the target value to 2 if the first column starts with 'luad'
data.loc[data.iloc[:, 0].str.startswith('luad'), 'target'] = 2

display(data)
```

```python


breast_cancer = data[data.iloc[:, 0].str.startswith('brca')]

display(breast_cancer)

prostate_cancer = data[data.iloc[:, 0].str.startswith('prad')]

display(prostate_cancer)

lung_cancer = data[data.iloc[:, 0].str.startswith('luad')]

display(lung_cancer)


```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from math import sqrt
from sklearn.model_selection import GridSearchCV


# drop the first column 
# x = the data frame with all the rows and all the columns except the first one
# y = the target column
x = data.iloc[:, 1:-1]
y = data['target']

#x_train = the data we use to train the model
#x_test = the data we use to test the model
#y_train = the target values we use to train the model
#y_test = the target values we use to test the model

# test_size = 0.2 means that 20% of the data will be used for testing and 80% will be used for training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=69)

# split the data into training and testing data for each cancer type
# split the data to avoid overfitting of a single cancer type (in our case, breast cancer)

x_breast_cancer = breast_cancer.iloc[:, 1:-1]
y_breast_cancer = breast_cancer['target']
x_train_breast_cancer, x_test_breast_cancer, y_train_breast_cancer, y_test_breast_cancer = train_test_split(x_breast_cancer, y_breast_cancer, test_size=0.2, random_state=69)

x_prostate_cancer = prostate_cancer.iloc[:, 1:-1]
y_prostate_cancer = prostate_cancer['target']
x_train_prostate_cancer, x_test_prostate_cancer, y_train_prostate_cancer, y_test_prostate_cancer = train_test_split(x_prostate_cancer, y_prostate_cancer, test_size=0.2, random_state=69)

x_lung_cancer = lung_cancer.iloc[:, 1:-1]
y_lung_cancer = lung_cancer['target']
x_train_lung_cancer, x_test_lung_cancer, y_train_lung_cancer, y_test_lung_cancer = train_test_split(x_lung_cancer, y_lung_cancer, test_size=0.2, random_state=69)

# combine the training and testing data for each cancer type back into one
x_train_split = pd.concat([x_train_breast_cancer, x_train_prostate_cancer, x_train_lung_cancer])
x_test_split = pd.concat([x_test_breast_cancer, x_test_prostate_cancer, x_test_lung_cancer])
y_train_split = pd.concat([y_train_breast_cancer, y_train_prostate_cancer, y_train_lung_cancer])
y_test_split = pd.concat([y_test_breast_cancer, y_test_prostate_cancer, y_test_lung_cancer])



percepNormal = Perceptron()
percepSplit = Perceptron()

# Train the classifier
percepNormal.fit(x_train, y_train)
percepSplit.fit(x_train_split, y_train_split)


# Predict the labels for the test set
y_pred = percepNormal.predict(x_test)
y_pred_split = percepSplit.predict(x_test_split)

# Print the accuracy of the classifier as a percentage
accuracy = percepNormal.score(x_test, y_test) * 100
print(f'Accuracy: {accuracy:.2f}%')

accuracy_split = percepSplit.score(x_test_split, y_test_split) * 100
print(f'Accuracy w/ Split: {accuracy_split:.2f}%')

# accuracy for no split data is higher than the accuracy for split data becauseeeee idk.

# tried using this function and it just took 5 ever to run
# gridSearch = GridSearchCV(Perceptron(), { 'max_iter': [1000, 10000], 'tol': [1e-3, 1e-4, 1e-5], 'eta0': [1, 0.1, 0.01, 0.001, 0.0001] }, cv=5)
# gridSearch.fit(x_train, y_train)
# print(gridSearch.best_params_)
# print(gridSearch.best_score_)
# print(gridSearch.score(x_test, y_test))

```
