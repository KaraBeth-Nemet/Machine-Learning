```python
import pandas as pd

data = pd.read_csv('dataset.csv')

display(data)
```

```python
# og number of rows 
print(data.shape)

# drop NAN values
# Pandas function : "Drop the rows where at least one element is missing."
data = data.dropna()
#print data sizes so we can see if we dropped any rows 
print(data.shape)

# drop duplicates
# Pandas function : "By default, it removes duplicate rows based on all columns."
data = data.drop_duplicates()
# print data sizes so we can see if we dropped any rows
print(data.shape)
```

```python
# find all the different types of samples and seperate them into their own dataframes

# print(data.iloc[:, 0])


breast_cancer = data[data.iloc[:, 0].str.startswith('brca')]

display(breast_cancer)

prostate_cancer = data[data.iloc[:, 0].str.startswith('prad')]

display(prostate_cancer)

lung_cancer = data[data.iloc[:, 0].str.startswith('luad')]

display(lung_cancer)


```
